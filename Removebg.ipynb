```python
import os
from rembg import remove
from PIL import Image
import io

# Define the input and output base directories
input_base_dir = r"D:\Indoor Plant\Orginal_Data\Data"
output_base_dir = r"D:\Indoor Plant\NonBackground_Image"

# Ensure the output base directory exists
os.makedirs(output_base_dir, exist_ok=True)

# Process each class folder in the dataset
for class_folder in os.listdir(input_base_dir):
    class_input_dir = os.path.join(input_base_dir, class_folder)
    
    # Check if it is a directory (class folder)
    if os.path.isdir(class_input_dir):
        # Create corresponding class folder in the output directory
        class_output_dir = os.path.join(output_base_dir, class_folder)
        os.makedirs(class_output_dir, exist_ok=True)

        # Process each image in the class folder
        for image_file in os.listdir(class_input_dir):
            if image_file.endswith(('png', 'jpg', 'jpeg')):
                input_path = os.path.join(class_input_dir, image_file)
                
                # Ensure the output file is saved with .png extension
                output_file_name = os.path.splitext(image_file)[0] + '.png'
                output_path = os.path.join(class_output_dir, output_file_name)

                # Open the image file
                with open(input_path, 'rb') as f:
                    input_image = f.read()

                # Remove the background
                try:
                    result = remove(input_image)  # This should return bytes
                    
                    # Convert bytes to an image object
                    result_image = Image.open(io.BytesIO(result))
                    
                    # Save the resulting image as .png
                    result_image.save(output_path, 'PNG')
                    print(f"Processed {image_file} from {class_folder} and saved to {output_path}")
                except Exception as e:
                    print(f"Failed to process {image_file} from {class_folder}: {e}")
```
